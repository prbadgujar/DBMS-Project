```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score


# Function to train model and compare results for a given category
def compare_sales(category_name):
    # Filter data for the specified category
    category_data = data[data['Category'] == category_name]

    # Features and target
    X = category_data[[ 'Order Month','Order Year']]  # Include additional features as necessary
    y = category_data['Sales']

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
    # Step 8: Feature Scaling (optional, improves performance for some models)
  #  scaler = StandardScaler()
   # X_train_scaled = scaler.fit_transform(X_train)
    #X_test_scaled = scaler.transform(X_test)
    
   # Build and train the Random Forest model
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)                          #removed X_train_scaled
    

    # Predict sales
    y_pred = model.predict(X_test)

    
    # Calculate R score
    r2 = r2_score(y_test, y_pred)

    # Print the R score
    print(f"R score: {r2:.2f}")

    # Prepare a DataFrame for comparison
    comparison_df = pd.DataFrame({
        'Month_Name': X_test['Order Month'].map({1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May',
                                           6: 'June', 7: 'July', 8: 'August', 9: 'September',
                                           10: 'October', 11: 'November', 12: 'December'}),
        'Sales': y_test,
        'Predicted_Sales': y_pred
    })

    # Group by month for an aggregated view
    comparison_df = comparison_df.groupby('Month_Name').agg({'Sales': 'sum', 'Predicted_Sales': 'sum'}).reset_index()


 # Sort the months in chronological order
    month_order = ['January', 'February', 'March', 'April', 'May', 'June',
                   'July', 'August', 'September', 'October', 'November', 'December']
    comparison_df['Month_Name'] = pd.Categorical(comparison_df['Month_Name'], categories=month_order, ordered=True)
    comparison_df = comparison_df.sort_values('Month_Name')




    
    # Display the comparison
    print(f"\nComparison of Historical and Future Sales for {category_name}:")
    print(comparison_df)

    # Plotting the comparison
    plt.figure(figsize=(10, 6))
    plt.plot(comparison_df['Month_Name'], comparison_df['Sales'], label='Actual Sales', marker='o')
    plt.plot(comparison_df['Month_Name'], comparison_df['Predicted_Sales'], label='Predicted Sales', marker='o')
    plt.xlabel('Month')
    plt.ylabel('Sales')
    plt.title(f'Comparison of Historical and Future Sales for {category_name}')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Perform the comparison for each category
compare_sales('Furniture')
compare_sales('Office Supplies')
compare_sales('Technology')
```
